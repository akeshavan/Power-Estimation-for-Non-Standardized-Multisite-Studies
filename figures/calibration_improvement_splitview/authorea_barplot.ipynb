```python
%pylab inline
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("LOO_calibration_effect.csv",index_col=0)
df.head()
```

```python
def paper_barplot(df):
    X = df.values
    fig,ax = subplots(1,2,figsize=(12,6))
    subplots_adjust(wspace=.3)
    N = X.shape[0]
    yax = arange(N)
    yax_labels = df.index
    before = X[:,0]
    after = X[:,1]

    wi_before = X[:,2]
    wi_after = X[:,3]

    width= 0.2
    space = 0.05
    sns.set_style("white")
    ax[0].spines["top"].set_visible(False)
    ax[0].spines["right"].set_visible(False)
    ax[0].spines["bottom"].set_visible(False)
    ax[0].spines["left"].set_visible(False)

    ax[1].spines["top"].set_visible(False)
    ax[1].spines["right"].set_visible(False)
    ax[1].spines["bottom"].set_visible(False)
    ax[1].spines["left"].set_visible(False)

    colors =  sns.color_palette("cubehelix",4)
    #colors = [0,0,0,0,"black","gray"]
    ax[0].barh(yax[:5],after[:5],height=width,label="BW After Calibration",color=colors[0],edgecolor="none")
    ax[0].barh(yax[:5]+width,before[:5],height=width,label="BW Before Calibration",color=colors[1],edgecolor="none")
    ax[0].barh(yax[:5]+space+2*width,wi_after[:5],height=width,label="WI After Calibration",  color=colors[2],edgecolor="none")
    ax[0].barh(yax[:5]+space+3*width,wi_before[:5],height=width,label="WI Before Calibration", color=colors[3],edgecolor="none")


    ax[1].barh(yax[:5]+space+3*width,wi_before[5:],height=width,label="WI Before Calibration", color=colors[3],edgecolor="none")
    ax[1].barh(yax[:5]+space+2*width,wi_after[5:],height=width,label="WI After Calibration", color=colors[2],edgecolor="none")
    ax[1].barh(yax[:5]+width,before[5:],height=width,label="BW Before Calibration",color=colors[1],edgecolor="none")
    ax[1].barh(yax[:5],after[5:],height=width,label="BW After Calibration",color=colors[0],edgecolor="none")

    roimap = {"EstimatedTotalIntraCranialVol":"eTIV",'Caudate':"Caud",
     'Amygdala':"Amyg",
     'Hippocampus':"Hipp",
     'Thalamus':"Thal",'Lateral-Ventricle':"LV",
     'CorticalWhiteMatterVol':"cWMV",
     'CortexVol':"cVol",
     'SubCortGrayVol':"scGMV",
     'TotalGrayVol':"GMV"}

    ax[0].set_yticklabels([roimap[z] for z in yax_labels[:5]],fontsize=16);
    ax[0].set_yticks(yax[:5]+2*width);
    ax[0].set_xticks([])



    ax[1].set_yticklabels([roimap[z] for z in yax_labels[5:]],fontsize=16);
    ax[1].set_yticks(yax[:5]+2*width);
    ax[1].set_xticks([])

    ax[1].legend(fontsize=12,bbox_to_anchor=(1.8, .6))
    ax[0].set_axisbelow(False)
    #ax.grid(zorder=6,color="white")

    for i,n in enumerate(after[:5]):
        ax[0].text(n*1.015,i+.03,"%.2f"%n,fontsize=12)

    for i,n in enumerate(before[:5]):
        ax[0].text(n*1.015,i+width*1.3,"%.2f"%n,fontsize=12)

    for i,n in enumerate(wi_after[:5]):
        ax[0].text(n*1.015,i+2*width*1.3-space,"%.2f"%n,fontsize=12)

    for i,n in enumerate(wi_before[:5]):
        ax[0].text(n*1.015,i+3*width*1.3-space,"%.2f"%n,fontsize=12)


    for i,n in enumerate(after[5:]):
        ax[1].text(n*1.015,i+.03,"%.2f"%n,fontsize=12)

    for i,n in enumerate(before[5:]):
        ax[1].text(n*1.015,i+width*1.3,"%.2f"%n,fontsize=12)

    for i,n in enumerate(wi_after[5:]):
        ax[1].text(n*1.015,i+2*width*1.3-space,"%.2f"%n,fontsize=12)

    for i,n in enumerate(wi_before[5:]):
        ax[1].text(n*1.015,i+3*width*1.3-space,"%.2f"%n,fontsize=12)

    suptitle("Absolute Agreement Within (WI) and Between (BW) Sites",fontsize=20)
    #savefig("calibration_improvement_splitview_revision.png",bbox_inches="tight",dpi=300)
```

```python
paper_barplot(df)
```
